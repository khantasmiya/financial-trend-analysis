```python
import pandas as pd
```

```python
data = {
    'Company': ['Apple', 'Apple', 'Apple', 'Samsung', 'Samsung', 'Samsung'],
    'Year': [2021, 2022, 2023, 2021, 2022, 2023],
    'Total Revenue': [300000, 320000, 340000, 280000, 290000, 310000],
    'Net Income': [50000, 52000, 55000, 46000, 47000, 49000]
}

df = pd.DataFrame(data)
df
```

```python
# Sort data by company and year
df = df.sort_values(by=['Company', 'Year'])

# Calculate Revenue Growth (%)
df['Revenue Growth (%)'] = df.groupby('Company')['Total Revenue'].pct_change() * 100

# Calculate Net Income Growth (%)
df['Net Income Growth (%)'] = df.groupby('Company')['Net Income'].pct_change() * 100

df
```

# 📊 Financial Trend Analysis using pandas

This report analyzes the financial performance of two companies (Apple and Samsung) over 3 years using Python and pandas.

## Methodology
- Simulated financial data was created.
- The percentage change in revenue and net income was calculated using `pct_change()` grouped by company.
- The goal was to identify year-over-year growth trends.

## Key Observations
- Apple shows steady revenue growth: 6.67% in 2022 and 6.25% in 2023.
- Samsung also shows positive trends: 3.57% in 2022 and 6.90% in 2023.
- Both companies have consistent net income growth, indicating financial stability.

## Conclusion
Pandas makes it easy to perform grouped calculations and analyze company trends over time.
This basic analysis can be expanded with more companies and additional financial metrics in the future.
